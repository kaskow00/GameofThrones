### Name the files battles, deaths and predict then append them all in one under the name "data"

```python
A = [
    "battles"
]
data = {}
for f in A:
    battles = pd.read_csv("./Game of Thrones/battles.csv")
print (f)


B = [
    "deaths"
]
data = {}
for f in B:
    deaths = pd.read_csv("./Game of Thrones/character-deaths.csv")
print (f)


C = [
    "predict"
]
data = {}
for f in C:
    predict= pd.read_csv("./Game of Thrones/character-predictions.csv")
print (f)


index = {'A': battles, 'B': deaths, 'C': predict}
data = pd.concat(index)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.formula.api as smf

data = pd.concat(index)
data.head(3)
lm = smf.ols(formula='culture ~ popularity', data=data).fit()
print (lm.params)
```
